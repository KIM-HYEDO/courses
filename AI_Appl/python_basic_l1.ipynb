* #### <u>내가 하기 싫은 일을 남이(기계가) 대신하게 한다</u>  
  
  
* 프로그래밍을 잘 한다는 것?
    + 기계가 잘 알아듣게 이야기 한다(프로그래밍 `언어` 라 불리는 EU)
  
  
* 고급 수준의 프로그래머라면?
    + 기계가 고급진(부가가치가 높은)일을 잘 하게 할 수 있다      
  
  
* 영어를 잘 하려면 미국에 가서 영어로 의사소통을 많이 해야 하고,
    + 프로그래밍을 잘 하려면 일단 컴퓨터를 켜서, `프로그래밍을 많이 해봐야 한다.` 
  
  
* 당연히 시키는 `일`과 관련한 배경지식도 필수
    + 일을 잘 아는 사람이 사람이 잘 시킨다  
  
  
* 결국 세상은 `남에게 일을 시킬 수 있는 사람`과 `남이 시키는 일만 할 수 있는 사람`으로 양극화 된다.
    + 프로그래밍을 할 줄 알아야 일을 시키며 살 수 있다.
    + 시키는 일만 하면 되는 업무는 앞으로 AI가 대체할 가능성이 높다.
    

### 1.2 자동화의 예시 1 - 코스피 종가 데이터 자동으로 가져와 그래프 그리기 

```python
'''
코스피 종가 데이터를 자동으로 가져와 그래프로 그려주는 스크립트 입니다.
pandas_datareader, yfinance, matplotlib 라이브러리를 설치해줘야 합니다.
* 설치방법 (터미널 기준)
pip install pandas_datareder
pip install yfinance
pip install matplotlib
'''

from pandas_datareader import data as pdr  
from matplotlib import pyplot as plt
import yfinance as yf
yf.pdr_override()

df = pdr.get_data_yahoo("^KS11") 
y_values = df['Adj Close']
x_values = range(len(y_values))

plt.plot(x_values,y_values)
plt.show()
```

- 몇 줄의 스크립트로 이런 자동화가 가능하다.

### 1.3 자동화의 예시 2 - 다음 헤드라인 뉴스 자동으로 가져오기

```python
''' 
다음 헤드라인 뉴스를 자동으로 가져오는 스크립트 입니다.
별도 라이브러리를 설치하지 않아도 실행 가능합니다.
'''
import urllib.request
import html.parser
u = 'http://www.daum.net'
f = urllib.request.urlopen(u)
contents = str(f.read().decode('utf-8'))
f.close()
i,k = 0,0

# infinite loop
while True:
    k = k + 1
    i = contents.find('뉴스1탭_텍스트소재', i+1)
    if i == -1 or k > 20:
        break
    i = contents.find('">', i+1)
    j = contents.find('</a>', i+1)
    data = contents[i+2:j]
    print(html.unescape(data))
```

- 기계에게 프로그래밍 언어로 원리만 잘 알려주면, 이런 것도 쉽게 할 수 있다.
