# 2. Python 기초 문법  
### 2.1 주석  
- 한줄 주석: `#` 기호로 시작하는 구문은 주석이 된다.

```python
# helloworld
print('hello world')
```

- 주석 블럭: `'''` 기호(작은 따옴표 세개) 또는 `"""` 기호(큰 따옴표 세개)로 둘러싼 블럭은 주석이 된다.

```python
'''
Function name: helloworld()
This function prints 'hello world'
'''
def helloworld():
    print('hello world')

helloworld()
```

### 2.2 들여쓰기와 코드블럭 
- 들여쓰기는 python에서 매우 중요한 역할을 하기 때문에 주의깊게 사용해야 한다. 
- 들여쓰기가 잘못된 경우 에러가 발생한다. `Indentation Error` 
- 들여쓰기는 공백(space) 개수와 탭(tab) 개수로 구분하나, 가독성과 일관성을 위해 탭 개수로 구분하는 것을 추천한다.
  
  
- __역할1__: 반복/제어 구문의 적용 범위를 나타낸다.  

```python
print('start')
for i in range(10):
    if i>5:
        print('i = %d > 5' % i)
        # if의 범위는 여기까지다.
    elif i<5:
        print('i = %d < 5' % i)
    else:
        print('i = %d = 5' % i)
    # for의 범위는 여기까지다.
print('end')
```

* __역할2__: 함수/클래스 정의 범위를 나타낸다.

```python
def function1():
    print('  function start')
    print('  function end')
# 여기가지 함수의 영역이다.

print('script start')
function1()
print('script end')
```

### 2.3 입출력하기  

- 입력 받기 `input()`
    + 화면을 통해 키보드로 입력을 받아 변수에 저장하게 해주는 함수이다.

```python
a = input()
print(a)
```

* 출력하기 `print()`
    + 원하는 내용을 화면에 출력할 수 있게 해주는 함수이다.
    + `' '` 사이에 출력할 내용의 표현식을 적는다.
    + 정해진 내용이 아닌, 변수에 따라 달라질 내용은 변수형에 따라 `%d`(정수) `%f`(실수) `%s`(문자열)을 넣는다.
        - `%.1f` (실수를 소숫점 첫째 자리까지만 표시)
    + `' '` 우측에 `%` 기호를 넣고 이어서 `( )` 안에 표현식에 대입할 변수들을 차례로 입력한다.

```python
i = 4
j = 4.5
k = 'abc'
print('i=%d j=%.1f k=%s' % (i,j,k))
```
